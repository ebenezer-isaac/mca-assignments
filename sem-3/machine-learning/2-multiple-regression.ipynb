```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
ocean = pd.read_csv('ocean.csv')
print(ocean.shape)
ocean.reset_index()
ocean['depth'].replace('', np.nan, inplace=True)
ocean['salinity'].replace('', np.nan, inplace=True)
ocean['temperature'].replace('', np.nan, inplace=True)
ocean.dropna(subset=['depth'], inplace=True)
ocean.dropna(subset=['salinity'], inplace=True)
ocean.dropna(subset=['temperature'], inplace=True)
print(ocean.shape)
ocean.head()
```

```python
ocean=ocean.astype(int)
```

```python
X = ocean.drop("temperature", axis=1)
X.head()
```

```python
Y = ocean['temperature']
Y.head()
```

```python
plt.plot(X['depth'],Y, color='blue', linewidth=1)
```

```python
plt.plot(X['salinity'],Y, color='blue', linewidth=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 17)
```

```python
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_prediction =  LR.predict(x_test)
y_prediction
```

```python
score=LR.score(x_test,y_test)
print("r2 socre is ",score)
print("mean_sqrd_error is==",mean_squared_error(y_test,y_prediction))
print("root_mean_squared error of is==",np.sqrt(mean_squared_error(y_test,y_prediction)))
```

```python
# Plot outputs
plt.scatter(y_test,y_prediction,  color='black')
plt.xticks(())
plt.yticks(())
plt.show()
```
