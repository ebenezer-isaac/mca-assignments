**Task  1**: Implement  the  game  "Guess theNumber!".  In  this  game,  the  computer  randomly chooses a number (e.g., between 1 and 50) and asks the player to guess the correct number. After each guess, the computer informs the user if the guess was correct, too high, or too low. The program should also inform the user how many guesses the user needed. After a certain limit (e.g., five guesses), the program stops and informs the userof the correct number.

```python
import random
print("Guess the Number between 0 and 50")
number = random.randint(0, 50)
trials=0
while trials<5:
  guess = int(input("Enter your guess : "))
  if guess == number:
    break
  elif guess<number:
    trials=trials+1
    print("Guess is too low")
  elif guess>number:
    print("Guess is too high")
    trials=trials+1
print("Correct number was",number)
```

```python
print("Entering stock for the fruit market : ")
print("Enter fruit name, stock, price separated by space in respective order. Press enter to exit loop.")
choice = "0"
fruits = {}
while len(choice)>0:
  choice = input("Your input : ")
  if len(choice)>0:
    choice = choice.split(" ")
    if len(choice)==3:
      fruits[choice[0]] = {
          "price":float(choice[2]),
          "stock":int(choice[1])
      }
    else:
      print("Invalid Input")
  else:
    break
  
```

```python
print("Entering bill for the fruit market : ")
print("Enter fruit name, qty separated by space in respective order. Press enter to exit loop.")
amount=0
choice="0"
bill=""
while len(choice)>0:
  choice = input("Your input : ").lower()
  if len(choice)>0:
    choice = choice.split(" ")
    if len(choice)==2:
      name = choice[0]
      qty = int(choice[1])
      if(int(fruits[name]["stock"])<qty): 
        print(name.title()+" : Not Enough Stock")
      else:
        fruits[name]["stock"] = str(int(fruits[name]["stock"])-qty)
        price = (float(fruits[name]["price"])*qty)
        amount = amount+ price
        bill_text = name.title()+" : "+str(fruits[name]["price"])+" * "+str(qty)+" = "+str(price)
        bill = bill+bill_text+"\n"
        print(bill_text)
    else:
      print("Invalid Input")
total = "Bill amount : "+str(amount)
bill  = bill +total
print(total)
```

**Task 5** : Write  a  simple  Python  program  using  an  editor  on  your  laptop,  e.g.,  a  calculator,  a game, etc., but something different from above tasks The program should be interactive, i.e., ask the user for input, perform some processing, and present an output.

```python
print("Ceaser Cipher Encryptor Program : ")
shift = 0
limits = [[48,57],[65,90],[97,122]]

def checkLimits(ascii, shift, correction):
  for limit in limits:
    if ascii>=limit[0] and ascii<=limit[1]:
      ascii = ascii+shift
      while ascii<limit[0]:
        ascii = (limit[1]+correction-(limit[0]-ascii))
      while ascii>limit[1]:
        ascii = (limit[0]+correction+(ascii-limit[1]))
      return chr(ascii)

def shiftText(message, shift):
  result = ""
  correction = -1 if shift>0 else 1
  for i in range(0,len(message)):
    ascii = ord(message[i])
    result  = result + checkLimits(ascii, shift, correction)
  return result  

while True:
  choice = int(input("Choose operation from the following:\n 1. Change Shift Key\n 2. Encrypt Message\n 3. Decrypt Message\n 4. Exit\n Your Choice : "))
  if choice==1:
    shift = int(input("Enter new shift value : "))
    print("New Shift Key : "+str(shift))
  elif choice ==2:
    message = input("Enter message to encrypt : ")
    result = shiftText(message,shift)
    print("Encrypted Message :"+str(result))
  elif choice ==3:
    message = input("Enter message to decrypt : ")
    result = shiftText(message,shift*-1)
    print("Decrypted Message :"+str(result))
  elif choice ==4:
    break
```
